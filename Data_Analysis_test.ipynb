##1. How much heavier is the age group with the highest average weight than the age group with the lowest weight?
answer in % please

```python
# Calculate average weight for each age group
average_weight_by_age = df_base.groupby('age')['weight'].mean()

# Identify age group with the highest average weight
highest_weight_age_group = average_weight_by_age.idxmax()

# Identify age group with the lowest average weight
lowest_weight_age_group = average_weight_by_age.idxmin()

# Calculate the difference in weight between the highest and lowest average weight age groups
weight_difference = average_weight_by_age[highest_weight_age_group] - average_weight_by_age[lowest_weight_age_group]

# Print the weight difference
print("The age group with the highest average weight is", highest_weight_age_group, "years old.")
print("The age group with the lowest weight is", lowest_weight_age_group, "years old.")
print("The age group with the highest average weight is", weight_difference, "units heavier than the age group with the lowest weight.")
```

```python
# Grouping the data by 'age' and calculating average weight
age_group_weights = df_base.groupby('age')['weight'].mean()

# Finding the age group with the highest average weight and the age group with the lowest weight
max_weight_age_group = age_group_weights.idxmax()
min_weight_age_group = age_group_weights.idxmin()

# Calculating the percentage difference in weight
percentage_difference = (age_group_weights[max_weight_age_group] - age_group_weights[min_weight_age_group]) / age_group_weights[min_weight_age_group] * 100

# Printing the result
print("The age group with the highest average weight is {:.2f}% heavier than the age group with the lowest weight.".format(percentage_difference))
```

## 2. Do people over 50 have higher cholesterol levels than the rest?

```python
# Creating subsets for individuals over 50 and below or equal to 50
over_50 = df_base[df_base['age'] > 50]
below_50 = df_base[df_base['age'] <= 50]

# Calculating the average cholesterol levels for each subset
average_cholesterol_over_50 = over_50['cholesterol'].mean()
average_cholesterol_below_50 = below_50['cholesterol'].mean()

# Comparing the average cholesterol levels
if average_cholesterol_over_50 > average_cholesterol_below_50:
    print("People over 50 have higher cholesterol levels than the rest.")
else:
    print("People over 50 do not have higher cholesterol levels than the rest.")
```

```python
# Calculate the average cholesterol levels for individuals over 50
avg_cholesterol_over_50 = df_base[df_base['age'] > 50]['cholesterol'].mean()

# Calculate the average cholesterol levels for individuals below 50
avg_cholesterol_below_50 = df_base[df_base['age'] <= 50]['cholesterol'].mean()

# Calculate the percentage difference
percentage_difference = (avg_cholesterol_over_50 - avg_cholesterol_below_50) / avg_cholesterol_below_50 * 100

# Print the result
if percentage_difference > 0:
    print("People over 50 have {:.2f}% higher cholesterol levels than the rest.".format(percentage_difference))
else:
    print("People over 50 have {:.2f}% lower cholesterol levels than the rest.".format(abs(percentage_difference)))
```

## 3. are men more likely to be a smoker than women? note the data contains information to identify gender IDs!

```python
# Calculate the total number of smokers for each gender
male_smokers = df_base[df_base['gender'] == 2]['smoke'].sum()
female_smokers = df_base[df_base['gender'] == 1]['smoke'].sum()

# Calculate the total number of individuals for each gender
male_count = len(df_base[df_base['gender'] == 2])
female_count = len(df_base[df_base['gender'] == 1])

# Calculate the percentage of smokers for each gender
male_smoking_percentage = (male_smokers / male_count) * 100
female_smoking_percentage = (female_smokers / female_count) * 100

# Calculate the ratio of male smoking percentage to female smoking percentage
smoking_ratio = male_smoking_percentage / female_smoking_percentage

# Print the result
if smoking_ratio > 1:
    print("Yes, men are {:.2f} times more likely to be smokers than women.".format(smoking_ratio))
elif smoking_ratio < 1:
    print("No, women are {:.2f} times more likely to be smokers than men.".format(1 / smoking_ratio))
else:
    print("The smoking prevalence is the same for both men and women.")
```

## 5. which two features have the highest spearman rank correlation?

```python
# Calculate the Spearman rank correlation matrix
spearman_corr_matrix = df_base.corr(method='spearman')

# Find the pair of features with the highest correlation
highest_corr_pair = spearman_corr_matrix.unstack().sort_values(ascending=False).drop_duplicates().head(2)

# Print the result
feature1, feature2 = highest_corr_pair.index
correlation = highest_corr_pair.values[0]
print("The two features with the highest Spearman rank correlation are '{}' and '{}', with a correlation coefficient of {:.2f}.".format(feature1, feature2, correlation))
```

## 6. what % of people are more than 2 standard deviation far from the average height?

```python


# Calculate the average height
average_height = df_base['height'].mean()

# Calculate the standard deviation of height
std_deviation = df_base['height'].std()

# Calculate the z-scores for each individual's height
z_scores = (df_base['height'] - average_height) / std_deviation

# Calculate the percentage of people more than 2 standard deviations away
percentage = (np.abs(z_scores) > 2).mean() * 100

# Print the result
print("{:.2f}% of people are more than 2 standard deviations away from the average height.".format(percentage))
```

## 6. What % of the population over 50 years old consume alcohol?
Also use the df_alco and merge the datasets on ID. Ignore those persons where we have no alcohol consumption information

```python
# Create a new 'id' column with numeric IDs
df_alco['id'] = range(len(df_alco))

# Rename the 'id;alco' column to 'alco'
df_alco.rename(columns={'id;alco': 'alco'}, inplace=True)

# # # Print the updated dataframe
# # print(df_alco)
# # Set the 'id' column as the index
# df_alco.set_index('id', inplace=True)

# # Print the updated dataframe
# print(df_alco)




# Merge df_alco and df_base on 'id' column
merged_df = df_base.merge(df_alco, on='id', how='inner')

# Filter data for individuals over 50 years old
over_50 = merged_df[merged_df['age'] > 50]

# Calculate the percentage of population over 50 that consumes alcohol
percentage = (len(over_50) / len(df_base[df_base['age'] > 50])) * 100

# Print the percentage
print(f"The percentage of the population over 50 years old that consumes alcohol is: {percentage:.2f}%")
```

## 6. What is the 95% confidence level?

```python
import scipy.stats as stats

# Number of individuals over 50 who consume alcohol
num_alcohol_consumers = len(over_50)

# Total number of individuals over 50
total_over_50 = len(df_base[df_base['age'] > 50])

# Calculate the sample proportion
sample_proportion = num_alcohol_consumers / total_over_50

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / total_over_50)

# Calculate the z-score for 95% confidence (two-tailed)
z_score = stats.norm.ppf(0.975)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

# Print the 95% confidence interval
print(f"The 95% confidence interval for the percentage of the population over 50 years old that consumes alcohol is: [{lower_bound:.2%}, {upper_bound:.2%}]")

```
